## Parametric curves

```python
# parametric circle
var('t')
x(t)=cos(t)
y(t)=sin(t)
circle = parametric_plot((x,y), (t,0,2*pi), color='blue', thickness=2)
def tangent_line(t0, length=0.5):
    dx = diff(x,t)(t0)
    dy = diff(y,t)(t0)
    xt = x(t0)
    yt = y(t0)
    return line([(xt-length*dx, yt-length*dy),(xt+length*dx, yt+length*dy)], color='red')
tangents = sum([tangent_line(t0) for t0 in [0,pi/2,pi,3*pi/2]])

show(circle+tangents, aspect_ratio=1, title='Circle with tangents')
    
```

```python
# parametric spiral curve
var('t')
x(t) = t*cos(t)
y(t) = t*sin(t)
spiral = parametric_plot((x,y), (t,0,6*pi), color='turquoise', thickness=2, alpha=0.4)
def tangent_line(t0, length=1):
    dx = diff(x,t)(t0)
    dy = diff(y,t)(t0)
    xt = x(t0)
    yt = y(t0)
    norm = sqrt(dx^2+dy^2)
    dx /= norm
    dy /= norm
    return line([(xt-length*dx, yt-length*dy), (xt+length*dx, yt+length*dy)], color='purple', linestyle='--')

tangent_lines2 = sum([tangent_line(t0) for t0 in [pi,2*pi,3*pi,4*pi,5*pi]])
show(spiral+tangent_lines2, aspect_ratio=1, title='Spiral curve with tangents')
```
